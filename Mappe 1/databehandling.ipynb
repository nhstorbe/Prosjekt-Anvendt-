```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
#henter data fra json filen
with open('weather.json', 'r') as file:
    weather_data = json.load(file)

timeseries = weather_data["properties"]["timeseries"]

#lager en liste med tid og temperatur
weather_list = [
    {"time": entry["time"], "temperature": entry["data"]["instant"]["details"]["air_temperature"]}
    for entry in timeseries
]

df = pd.DataFrame(weather_list)  
print(df.head())
```

```python
plt.figure(figsize=(20,10))

df["time"] = pd.to_datetime(df["time"]) 
plt.plot(df["time"], df["temperature"], label="Temperatur")

plt.xlabel("Tid")
plt.ylabel("Temperatur (°C)")
plt.title("Værdata fra Yr.no - Cecilekruna 1717 moh.")

plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.show()

#DETTE ER IKKE RETT, MÅ SE PÅ VÆRDATA FILEN OM VI HAR HENTET FRA RETT STED FOR DET ER IKKE 29 GRADER I STRYN
```

```python
#henter data fra json filen
with open('weather.json', 'r') as file:
    weather_data = json.load(file)

timeseries = weather_data["properties"]["timeseries"]

#lager en liste med tid og vindkast
weather_list = [
    {"time": entry["time"], "wind_speed": entry["data"]["instant"]["details"]["wind_speed"]}
    for entry in timeseries
]

df = pd.DataFrame(weather_list)  
print(df.head())


plt.figure(figsize=(20,10))

df["time"] = pd.to_datetime(df["time"]) 
plt.plot(df["time"], df["wind_speed"], label="Vindkast")

plt.xlabel("Tid")
plt.ylabel("Vindkast (m/s)")
plt.title("Værdata fra Yr.no - Cecilekruna 1717 moh.")

plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
